```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r"C:\Users\Hp\Desktop\M Lbin\Heart Prediction Quantum Dataset.csv")
```

Using Random Forest Classifier

```python
X=df.drop('HeartDisease',axis=1)
df['HeartDisease'].value_counts()
```

```python
Y=df['HeartDisease'].apply(lambda y_value:1 if y_value>0 else 0)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(Y.shape,Y_test.shape,Y_train.shape)
```

```python
Y
```

```python
print(Y_train.value_counts())
```

```python
model=RandomForestClassifier(random_state=42)
model.fit(X_train,Y_train)
```

```python
X_test_prediction=model.predict(X_test)
```

```python
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
test_data_accuracy
```

```python
input_data=(56,1,180,214,65,9.445544668)
input_data_as_array=np.asarray(input_data)
```

```python
print("Original shape:", input_data_as_array.shape)

input_data_reshaped=input_data_as_array.reshape(1,-1)
prediction=model.predict(input_data_reshaped)
print(input_data_reshaped.shape)
print(prediction)
if (prediction[0]==1):
    print("Heart Disease")
else:
    print("safe")
```

Using Decission Tree Classifier

```python
X1=df.drop(['HeartDisease'],axis=1)
Y1=df['HeartDisease'].apply(lambda y_value:1 if y_value>0 else 0)
```

```python
X1_train,X1_test,Y1_train,Y1_test=train_test_split(X1,Y1,test_size=0.2,random_state=3)
print(Y1.shape,Y1_train.shape,Y1_test.shape)
model1=DecisionTreeClassifier()
model1.fit(X1_train,Y1_train)
print(Y1_train.value_counts)
```

```python
X1_test_prediction=model1.predict(X1_test)
test_data_accuracy1=accuracy_score(X1_test_prediction,Y1_test)
```

```python
print("Accuracy score:",test_data_accuracy1)
```

```python
input_data1=(40,1,121,255,96,9.683800175)
input_data_as_array1=np.asarray(input_data1)
input_data_reshaped1=input_data_as_array1.reshape(1,-1)
prediction1=model1.predict(input_data_reshaped1)
print(input_data_reshaped1.shape)
print(prediction1)
if(prediction1>0):
    print("Heart Disease")
else:
    print("safe")
```
